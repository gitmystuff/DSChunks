# Central Limit Theorem

```python
import numpy as np
import matplotlib.pyplot as plt

# law of large numbers
np.random.seed(1)
rolls = np.random.randint(1, 7, 500)
# print(rolls)
print(np.mean(rolls))

# central limit theorem
# means = [np.mean(np.random.randint(1, 7, 50)) for _ in range(1000)]
means = []
for i in range(1000):
    means.append(np.mean(np.random.randint(1, 7, 50)))
    
print(means[0:10])
plt.hist(means)
plt.show()
```
