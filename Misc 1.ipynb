# Law of Large Numbers

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

x = np.linspace(-3, 3, 100)
y = stats.norm.pdf(x, 0, 1)
plt.plot(x, y)
plt.show()
```

```python
np.random.seed(1)
sizes = []
for x in range(10, 20000, 200):
    sizes.append(x)

means = [np.mean(np.random.randn(size)) for size in sizes]

pos = 20
print(sizes[0:pos])
print(means[0:pos])
x = sizes[pos]
y = means[pos]

# plt.scatter(sizes, np.array(means))
plt.plot(sizes, np.array(means))
plt.plot(x, y, 'ro')
plt.show()
```

```python
means = [np.mean(np.random.randn(size) + 50) for size in sizes]
plt.plot(sizes, np.array(means))
plt.show()
```

```python
xaxis = np.arange(30, 70, 1)
yaxis = stats.norm.pdf(xaxis, 50, 5)
plt.plot(xaxis, yaxis)
plt.show()
```

```python
xaxis = np.arange(-40, 40, 1)
yaxis = stats.norm.pdf(xaxis, 0, 10)
plt.plot(xaxis, yaxis)
plt.show()
```
