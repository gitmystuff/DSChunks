# The Law of Large Numbers

In probability theory, the law of large numbers is a theorem that describes the outcome of performing the same experiment many times. The law states that as the number of trials increases, the average of the results will get closer and closer to the expected value. Essentially, it "guarantees stable long-term results for the averages of some random events".

For example, when flipping a fair coin, the expected value is 0.5 (if we code heads as 1 and tails as 0). Flipping the coin only a few times might result in an average far from 0.5 (e.g., three heads in a row gives an average of 1). However, flipping it thousands of times will result in an average that is extremely close to 0.5.

**History:** The concept was first identified by Italian mathematician Gerolamo Cardano in the 16th century, though he didn't provide a rigorous proof. The first person to develop a mathematical proof for the law was Swiss mathematician Jacob Bernoulli, whose theorem was published posthumously in 1713.

**Python Implementation**

The following script first creates a large dataset to serve as our "population." It then calculates the "true" mean of this population. Finally, it takes random samples of increasing sizes, calculates their means, and shows how these sample means converge on the true population mean.

```
import numpy as np
import matplotlib.pyplot as plt

# Set a seed for reproducibility, so the results are the same every time.
np.random.seed(42)

# 1. Create a large "population" of random data.
# We will generate 5000 random integers between 0 and 5000.
population_data = np.random.randint(0, 5000, 5000) #

# 2. Calculate the "true" mean of the population.
# This is our expected value.
population_mean = np.mean(population_data)
print(f"True Population Mean: {population_mean:.4f}")
print("-" * 30)

# 3. Define sample sizes and calculate the mean for each sample.
# We will take samples of increasing size from the population.
sample_sizes = [5, 50, 500, 1000, 2500, 4500, 5000]
sample_means = []

print("Calculating sample means:")
for size in sample_sizes:
    # Take a random sample from the population data.
    sample = np.random.choice(population_data, size)

    # Calculate the mean of the sample.
    mean = np.mean(sample)
    sample_means.append(mean)

    # Print the results, showing the difference from the true mean.
    print(f"Sample Size: {size:<5} | Sample Mean: {mean:<10.4f} | Difference: {abs(population_mean - mean):.4f}")

# 4. Visualize the results.
# This plot shows the sample means getting closer to the population mean as sample size increases.
plt.figure(figsize=(12, 7))
plt.plot(sample_sizes, sample_means, marker='o', linestyle='-', color='b', label='Sample Means')
plt.axhline(y=population_mean, color='r', linestyle='--', label=f'True Population Mean ({population_mean:.4f})')
plt.xlabel("Sample Size")
plt.ylabel("Mean")
plt.title("Demonstration of the Law of Large Numbers")
plt.legend()
plt.grid(True)
plt.show()
```
