**Git**

* **Version Control System:** At its core, Git is a tool that allows you to track changes to your files over time. Think of it like a super-powered "undo" button for your projects.
* **Local:** Git is installed and runs on your own computer. It maintains a history of every change made to your project in a special database called a repository (or "repo" for short).
* **Key Concepts:**
    * **Repository:** A folder that Git tracks. It contains all your project files and the entire history of changes.
    * **Commit:** A snapshot of your project at a specific point in time. You create a commit every time you want to save a set of changes.
    * **Branch:**  Like a separate line of development. You can create branches to work on new features or experiments without affecting the main project.
    * **Merge:** The process of combining changes from different branches.

## Git Workflow

**1. Prepare Repository**

* **Create a repository:**  Create a new repository on GitHub (e.g., "python-class-files").
* **README with Instructions:** In the README, provide clear instructions:
    * Fork the repository.
    * Clone the forked repository.
    * Create a new Python file (e.g., `firstname_lastname.py`).
    * Write a simple Python program in the file (you can provide a specific task or leave it open-ended).
    * Commit and push the changes.
    * Create a pull request.

**2. Contributor Workflow (Git Commands)**

* **Fork:** Contributors fork "python-class-files" repository.
* **Clone:**
   ```bash
   git clone <URL_of_forked_repo>
   ```
* **Create Python File:** Contributors use a text editor or IDE to create a new Python file (e.g., `john_doe.py`) inside the cloned repository folder.
* **Write Code:** They write their Python code in the file.
* **Stage Changes:**
   ```bash
   git add john_doe.py
   ```
* **Commit Changes:**
   ```bash
   git commit -m "Added my Python file"
   ```
* **Push Changes:**
   ```bash
   git push origin main
   ```
* **Pull Request:** Contributors create a pull request from their forked repository to your original repository.

**3. Accepting Pull Requests**

* **Review:** Review the pull requests to check the code.
* **Merge:**  Merge the pull requests individually or use a script/action for automatic merging if you have a large class.

**Advantages of this Approach:**

* **Individual work:** Each Contributor has their own file, reducing the chance of conflicts.
* **Code practice:** Contributors get to write and submit code.
* **Assessment:** You can easily review individual code submissions.

**Tips for Managing Many Files:**

* **Naming convention:**  Enforce a clear naming convention for the Python files (e.g., `firstname_lastname.py`) to keep things organized.
* **Folders (optional):** You might consider having Contributors create their files within folders named after them to further organize the repository.
* **Automated testing:** If you have the resources, consider setting up automated tests to check the submitted code for basic functionality.




## Syncing Repository

It's important to make sure everyone's local repositories are synchronized with the main repository on GitHub after you've merged all the changes.

**1. Explain the Concept**

Explain that Git is a distributed version control system. This means that everyone has their own complete copy of the repository on their computer. When you merge pull requests on GitHub, you're updating the main repository, but the changes aren't automatically reflected in their local copies. They need to explicitly fetch and merge those changes.

**2.  Git Commands to Update**

Here are the commands contributors need to run in their local repository (from the terminal):

* **Fetch changes:**
   ```bash
   git fetch origin
   ```
   This downloads the latest changes from the main repository on GitHub (`origin`) to their local repository. It doesn't automatically merge them yet.

* **Merge changes:**
   ```bash
   git merge origin/main
   ```
   This merges the fetched changes from the `origin/main` branch (the main branch on GitHub) into their current local branch (likely `main`).

**3. Alternative: Pull Command**

You can also teach them the `git pull` command, which combines `fetch` and `merge` in one step:

```bash
git pull origin main
```

This command fetches the latest changes from the `main` branch of the `origin` remote and merges them into the current local branch.

**4.  Visualizing with a Diagram**

It can be helpful to visualize this process with a simple diagram showing the local repository, the remote repository (origin), and the flow of changes with `fetch` and `merge`.

**5.  Practice and Repetition**

Encourage practicing these commands. Updating their local repository is a fundamental skill they'll need for any collaborative project using Git.

**Important Notes:**

* **Conflicts:** If a contributor has made local changes to the same file that was modified in the merged pull requests, they might encounter merge conflicts. They'll need to resolve these conflicts manually before they can update their repository.
* **Branching:** If contributors are working on different branches, they'll need to make sure they are on the correct branch before merging changes.
* **GUI tools:** If your contributors are using a Git GUI client (like GitHub Desktop), they can usually update their repository with a few clicks instead of using the command line.
