Market Basket Analysis

Association rule mining is a widely used technique for discovering relationships between variables in a large data set (e.g., items in a store that are commonly purchased together). The classic market basket analysis example using association rules is the "beer and diapers" rule. According to a data mining urban legend, a study of customer purchase behavior in a supermarket found that men often purchased beer and diapers together. After making this discovery, the managers strategically placed beer and diapers closer together on the shelves and saw a dramatic increase in sales. In addition to market basket analysis, association rules are also used in bioinformatics, web analytics, and several other fields.

https://madlib.apache.org/docs/latest/group__grp__assoc__rules.html

In another tactic, managers may distance milk and bread so that customers are forced to walk through more aisles and see more products.

Apriori: If you do something, do you do something else? What is this something and something else?

### Rules

x => y: when does x and y appear together; when you have x, will you also have y; when you buy x, will you buy y

* Count: The number of transactions that contain x and y
* Support: Percentage of times the items appear together. Transaction of x and y / all transactions 
* Confidence: Percentage of transactions where this rule is valid. The ratio of transactions that contain x and y to transactions that contain x. One could view this metric as  p(x | y) or confidence(x => y).
* Lift: The strength of the association. The ratio of observed support(x +> y) to the expected support of x and y , assuming x and y are independent. lift(x => y) = support(x => y) / (support(x) * support(y)), or $p(x \cap y) / (p(x) * p(y))$
* Conviction: How sure are we? conviction(x => y) = (1 - support(x => y)) / (1 - confidence(x => y))

```python
# super datascience center
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/gitmystuff/Datasets/main/Market_Basket_Optimisation.csv', header = None)
transactions = []
for i in range(0, 7501):
    transactions.append([str(df.values[i,j]) for j in range(0, 20)])
    
df.head()
```

```python
# build rules
from apyori import apriori

rules = apriori(transactions = transactions, 
                min_support = 0.003, 
                min_confidence = 0.2, 
                min_lift = 3, 
                min_length = 2, 
                max_length = 2)
```

```python
# display results
results = list(rules)

def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))

analysis = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
analysis.nlargest(n = 10, columns = 'Lift')
```

## ECLAT

Equivalence Class Clustering and Bottom-Up Lattice Traversal

https://towardsdatascience.com/the-eclat-algorithm-8ae3276d2d17

Unlike the apriori method, the ECLAT method is not based on the calculation of confidence and lift, therefore the ECLAT method is based on the calculation of the support conjunctions of the variables.

https://pypi.org/project/pyECLAT/

* ECLAT focuses on support
* Faster with fewer metrics

Eclat is an algorithm for finding frequent item sets in a transaction or database. It is one of the best methods of Association Rule Learning. Which means Eclat algorithm is used to generate frequent item sets in a database.

https://www.i2tutorials.com/machine-learning-tutorial/machine-learning-eclat-algorithm/

```python
# !pip install pyECLAT
```

```python
# https://hands-on.cloud/implementation-of-eclat-algorithm-using-python/
from pyECLAT import ECLAT

min_n_products = 2
min_support = .03
max_length = 2

analysis = ECLAT(data=df, verbose=True)
_, rule_supports = analysis.fit(min_support=.03,
                                           min_combination=2,
                                           max_combination=2)
```

```python
import pandas as pd

eclat_df = pd.DataFrame(rule_supports.items(), columns=['items', 'support'])
eclat_df.sort_values(by=['support'], ascending=False).head(10)
```
