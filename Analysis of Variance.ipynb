# Analysis of Variance (ANOVA)

### The Need for ANOVA

ANOVA is a statistical test used to determine if there are any statistically significant differences between the means of **three or more** independent (unrelated) groups.

#### The Problem with Multiple $t$-tests

If you have three groups (Group 1, Group 2, Group 3), you might be tempted to test all possible pairs using separate $t$-tests:

* $H_0: \mu_1 = \mu_2$
* $H_0: \mu_1 = \mu_3$
* $H_0: \mu_2 = \mu_3$

If you set the significance level ($\alpha$) for each individual test at $0.05$ (or a $95\%$ confidence level), the **overall confidence** that you have made *no Type I error* (false positive) across the entire experiment **decreases**.

**Overall Confidence Calculation:**
$$0.95 \times 0.95 \times 0.95 \approx 0.857$$
This means the **overall Type I error rate** (or $\alpha$ for the whole experiment) increases to about $1 - 0.857 = 0.143$, or $14.3\%$. ANOVA controls this overall error rate, keeping it at your desired level (e.g., $\alpha = 0.05$).

***

### The Null Hypothesis ($H_0$)

The standard **Null Hypothesis** for a one-way ANOVA is:
$$H_0: \mu_1 = \mu_2 = \mu_3 = \dots = \mu_k$$
(The means of all groups are equal.)

The **Alternative Hypothesis** ($H_a$) is:
$$H_a: \text{At least one group mean is different from the others.}$$

***

### How the $F$-Test Works in ANOVA

ANOVA works by calculating an **$F$-statistic** and comparing it to a **critical value** determined by the **degrees of freedom ($\text{df}$)**. The $F$-statistic is a ratio of two types of variance:

$$\text{F} = \frac{\text{Variance Between Groups (Signal)}}{\text{Variance Within Groups (Noise)}}$$

#### 1. Variance Between Groups ($\text{Mean Square Group, MS}_\text{G}$)

* This measures how much the **mean of each group** varies from the **overall grand mean** of all data points.
* It represents the variance *explained by* the different treatments or groups (the **signal**).

#### 2. Variance Within Groups ($\text{Mean Square Error, MS}_\text{E}$)

* This measures how much the **individual scores within each group** vary from their **own group's mean**.
* It represents the unexplained or random variance (the **noise** or error).

#### The $F$-Statistic Formula (Mean Squares)

$$F = \frac{\text{MS}_\text{G}}{\text{MS}_\text{E}}$$

#### The $F$-Statistic Formula (Sum of Squares and Degrees of Freedom)

The mean squares (MS) are calculated from the **Sum of Squares (SS)** divided by the corresponding **Degrees of Freedom ($\text{df}$)**:

$$F = \frac{\frac{\text{SS}_\text{G}}{\text{df}_{\text{groups}}}}{\frac{\text{SS}_\text{E}}{\text{df}_{\text{error}}}}$$

* **$\text{SS}_\text{G}$** (Sum of Squares Group) is the numerator's $\text{SS}$.
* **$\text{df}_{\text{groups}} = n_{\text{groups}} - 1$.** (Where $n_{\text{groups}}$ is the number of groups.)
* **$\text{SS}_\text{E}$** (Sum of Squares Error) is the denominator's $\text{SS}$.
* **$\text{df}_{\text{error}} = N_{\text{total}} - n_{\text{groups}}$.** (Where $N_{\text{total}}$ is the total number of items/observations across all groups.) *Note: The formula you provided, $(\mathbf{n}_{\text{rows}} - 1) * n_{\text{groups}}$, is a simplified, sometimes less precise way to express $\text{df}_{\text{error}}$ and assumes equal group sizes. The formula $N_{\text{total}} - n_{\text{groups}}$ is generally safer and more standard.*

***

### Understanding Variance and Degrees of Freedom

#### Variance (or Sample Variance)

The variance is fundamentally a measure of the average squared distance of each score from the mean:

$$\text{Sample Variance} = \frac{\sum(x - \bar{x})^2}{n - 1} = \frac{\text{SS}}{\text{df}}$$

#### Degrees of Freedom ($\text{df}$)

Degrees of freedom generally refer to the number of independent values or pieces of information used to estimate a parameter. In the context of variance, the denominator is $n-1$ because one degree of freedom is 'lost' when the sample mean ($\bar{x}$) is calculated, as it is needed to compute the deviations $(x - \bar{x})$.

***

### Assumptions of ANOVA

For the results of the ANOVA $F$-test to be reliable, the following assumptions about the data must be met:

1.  **Independent Samples:** The data points in one group must be independent of the data points in all other groups.
2.  **Continuous Dependent Variable:** The outcome variable (the thing being measured) must be measured on an interval or ratio scale.
3.  **Normally Distributed Errors (Residuals):** The residuals (the differences between the observed values and the group means) should be approximately normally distributed for each group.
4.  **Homogeneity of Variances (Homoscedasticity):** The population variances for each group must be approximately equal. (i.e., the "spread" of data in each group should be similar.)

## The $F$-Distribution Explained

The **$F$-distribution** is a crucial **continuous probability distribution** in inferential statistics. It primarily models the **ratio of two independent sample variances** and is fundamental to the **$F$-test** and **Analysis of Variance (ANOVA)**.

### Core Characteristics

The $F$-distribution is defined by the following properties:

* **Ratio of Variances:** It models the distribution of the ratio $\frac{s_1^2}{s_2^2}$ when two independent samples are drawn from normal populations.
* **Non-Negative:** Its values are always **non-negative** ($F \ge 0$) since it is a ratio of squared quantities (variances).
* **Shape:** It is **right-skewed**. The specific shape depends on the degrees of freedom.
* **Two Degrees of Freedom ($\text{df}$):** Unlike the $t$ or $\chi^2$ distributions, the $F$-distribution requires **two separate $\text{df}$ values** to define its shape:
    1.  **Numerator $\text{df}$ ($\nu_1$):** Associated with the variance in the numerator ($n_1 - 1$).
    2.  **Denominator $\text{df}$ ($\nu_2$):** Associated with the variance in the denominator ($n_2 - 1$).


***

## General Application: The $F$-Test for Variances

The foundational use of the $F$-distribution is to test if the variances of two populations are equal.

* **Question:** What is the probability that two independent samples come from populations with the **same variance** ($\sigma_1^2 = \sigma_2^2$)?
* **Principle:** If the two population variances are equal, the expected $F$-ratio of their sample variances ($\frac{s_1^2}{s_2^2}$) is close to **$1.0$**.
* **Usage:** This test (or related versions like Levene's or Bartlett's) is often used to check the **homogeneity of variances** assumption, which is required for tests like ANOVA.

***

## Primary Application: The $F$-Distribution in ANOVA

In ANOVA, the $F$-distribution is used to compare the means of **three or more groups** simultaneously. ANOVA cleverly reframes the test for equal means as a test for equal variances.

* **ANOVA Question:** What is the probability that three or more samples come from the same population (i.e., **have the same population mean**)?
* **The $F$-Statistic:** The $F$-statistic in ANOVA is calculated as the ratio of two independent variance estimates:
   $$F = \frac{\text{MS}_{\text{G}}}{\text{MS}_{\text{E}}}$$

| Term | What it Measures | Interpretation |
| :--- | :--- | :--- |
| $\text{MS}_{\text{G}}$ | **Variance Between Groups** (Effect + Error) | Estimates $\sigma^2$ only if $H_0$ is true. |
| $\text{MS}_{\text{E}}$ | **Variance Within Groups** (Error Only) | Always a good estimate of $\sigma^2$ (the error). |

### Interpretation of the $F$-Ratio

| Scenario | Hypothesis | Expected $F$-Value | Conclusion |
| :--- | :--- | :--- | :--- |
| **No Real Difference** | $H_0: \mu_1 = \mu_2 = \mu_3$ is **True** | $\text{MS}_{\text{G}} \approx \text{MS}_{\text{E}}$, so $F \approx 1.0$ | Group means are not significantly different. |
| **Real Difference** | $H_0$ is **False** | $\text{MS}_{\text{G}} > \text{MS}_{\text{E}}$, so $F \gg 1.0$ | Group means are significantly different. |

A **large $F$-value** indicates that the differences *between* the group means ($\text{MS}_{\text{G}}$) are much greater than the random variability *within* the groups ($\text{MS}_{\text{E}}$), suggesting a statistically significant difference between the population means.

***

## Contrast with $z$ and $t$ Distributions

The $F$-test is distinct from the $z$ and $t$ tests based on the number of means being compared:

* **$z$ and $t$ Tests:** Used for comparing **one or two means** (e.g., a sample mean to a population mean, or two sample means). The $t$-distribution is used when the population standard deviation is unknown.
* **ANOVA ($F$-Test):** The essential tool for comparing the means of **three or more groups** simultaneously, which controls the overall Type I error rate (the risk of falsely rejecting a true null hypothesis).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

data = {'Group1': [37, 60, 52, 43, 40, 52, 55, 39, 39, 23],
'Group2': [62, 27, 69, 64, 43, 54, 44, 31, 49, 57],
'Group3': [50, 63, 58, 54, 49, 52, 53, 43, 65, 43]}

anova_df = pd.DataFrame(data)
print(anova_df.head(10))

# sum of squares groups
g1_mean = anova_df['Group1'].mean()
g2_mean = anova_df['Group2'].mean()
g3_mean = anova_df['Group3'].mean()
tot_mean = sum(anova_df.mean())/anova_df.shape[1]
SSG1 = (g1_mean - tot_mean)**2
SSG2 = (g2_mean - tot_mean)**2
SSG3 = (g3_mean - tot_mean)**2

SSG = (SSG1 + SSG2 + SSG3) * anova_df.shape[0]
print(f'SSG = {SSG1} + {SSG2} + {SSG3} = {SSG1 + SSG2 + SSG3} * {anova_df.shape[0]} = {SSG}')
df_groups = anova_df.shape[1] - 1
print(f'df_groups = ngroups - 1 = {df_groups}')

# sum of squares error
SSE1 = [(i - g1_mean)**2 for i in anova_df['Group1']]
SSE2 = [(i - g2_mean)**2 for i in anova_df['Group2']]
SSE3 = [(i - g3_mean)**2 for i in anova_df['Group3']]

SSE = sum(SSE1 + SSE2 + SSE3)
print(f'SSE = SSE1 + SSE2 + SSE3 = {SSE}')
df_errors = (anova_df.shape[0] - 1) * anova_df.shape[1]
print(f'df_errors = (nrows - 1) * ngroups = {df_errors}')

# get f statistic
f_stat = (SSG/df_groups)/(SSE/df_errors)
print('f statistic: ', np.round(f_stat, 3))

cv = stats.f.ppf(q=1-.05, dfn=2, dfd=27)
print('critical value: ', np.round(cv, 3))

# visualize with plot
x = np.linspace(0, 10, 100)
y = stats.f.pdf(x, df_groups, df_errors)
plt.plot(x, y, label=f'df1 = {df_groups}, df2 = {df_errors}')
plt.axvline(x=f_stat, color='b', label='f statistic')
plt.axvline(x=cv, color='r', label='critical value')

plt.legend();
```

Since f is lest than $f_{critical}$, or 1.718 < 3.354, we fail to reject the hypothesis. We don't have enough evidence to show there are differences between the groups.

## Two-Way ANOVA

**Two-Way Analysis of Variance (ANOVA)** is a statistical test that evaluates how two **independent variables (factors)** simultaneously influence a single **continuous dependent variable (outcome)**.

The analysis is structured to test three specific null hypotheses:

### 1. The Main Effects and Interaction

| Hypothesis Test | Description |
| :--- | :--- |
| **Main Effect 1** (Factor A) | Is there a significant difference in the outcome across the levels of Factor A, *regardless* of Factor B's levels? |
| **Main Effect 2** (Factor B) | Is there a significant difference in the outcome across the levels of Factor B, *regardless* of Factor A's levels? |
| **Interaction Effect** (A $\times$ B) | Does the effect of Factor A **depend** on the specific level of Factor B (or vice versa)? |



***

## The Two-Way ANOVA with a Randomized Block Design

A special and common application of Two-Way ANOVA is the **Randomized Block Design**, often used in controlled experiments where one of the "factors" (the rows/blocks) is designated as a **nuisance variable**.

### 2. The Role of Blocks (Nuisance Variable)

In this design, the blocks are a **control mechanism**. They represent a known source of variation that could affect the outcome but is **not the primary research question**.

| Block Examples | Primary Goal |
| :--- | :--- |
| **Time of Day** (e.g., morning vs. afternoon) | To ensure time-based effects don't bias the main factor. |
| **Participant Age/Sex/Initial Ability** | To control for inherent differences between subjects. |
| **Location/Facility** (e.g., lab 1 vs. lab 2) | To account for environmental variations. |

### 3. The Goal: Improve Test Sensitivity

The primary statistical goal of blocking is to **isolate and statistically remove** the variance contributed by the blocks ($\text{SSB}$) from the total variation ($\text{SST}$).

* **Variance Isolation:** By accounting for Block Variation ($\text{SSB}$), the remaining unexplained variation, the **Sum of Squared Errors ($\text{SSE}$)**, is reduced.
* **Increased Power:** The F-statistic for the main effect of interest is calculated as the ratio of the Mean Square of the Group ($\text{MSG}$) to the Mean Square Error ($\text{MSE}$):
   $$F = \frac{\text{MSG}}{\text{MSE}}$$ (the denominator) effectively **increases the F-statistic** for the factor of interest. A larger F-statistic makes the test **more powerful**, increasing your chance of detecting a true effect of your group or treatment.

**In essence, blocking provides a cleaner, more sensitive measure of the true effect of the treatment groups by stripping away the known noise from the error term.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# set up two way df
data = {'Block': ['Block1', 'Block2', 'Block3' ], 'Group1': [8, 10, 12], 'Group2': [11, 12, 13]}
two_way_df = pd.DataFrame(data)
two_way_df.set_index('Block', inplace=True)
print(two_way_df.head())

# sum of squares groups
g1_mean = two_way_df['Group1'].mean()
g2_mean = two_way_df['Group2'].mean()
tot_mean = sum(two_way_df.mean())/two_way_df.shape[1]
SSG1 = (g1_mean - tot_mean)**2
SSG2 = (g2_mean - tot_mean)**2

SSG = (SSG1 + SSG2) * two_way_df.shape[0]
print(f'SSG = (SSG1 + SSG2) * nblocks = {SSG1 + SSG2} * {two_way_df.shape[0]} = {SSG}\n')

# add block means
two_way_df['BlockMeans'] = two_way_df.mean(axis=1)
print(two_way_df.head())

# total mean
tot_mean = sum(two_way_df.mean())/two_way_df.shape[1]
print('\nTotal Mean:', tot_mean)

# sum of squares block
SSB = sum([(i - tot_mean)**2 for i in two_way_df['BlockMeans']]) * 2
print(f'SSB = {SSB}')

# sum of squares total
SST = sum(sum([(i[1] - tot_mean)**2 for i in two_way_df.drop('BlockMeans', axis=1).iterrows()]))
print('SST =', SST)

# SSE
SSE = SST - SSG - SSB
print(f'SSE = SST - SSG - SSB = {SSE}')

# degrees of freedom
df_groups = two_way_df.drop('BlockMeans', axis=1).shape[1] - 1
print(f'df_groups = {df_groups}')

# df_errors changes to allow for SSB
df_errors = (two_way_df.drop('BlockMeans', axis=1).shape[0] - 1) * (two_way_df.drop('BlockMeans', axis=1).shape[1] - 1)
print(f'df_errors = {df_errors}')

f_stat = (SSG/df_groups)/(SSE/df_errors)
print(f'f statistic = {f_stat}')

cv = stats.f.ppf(1 - .05, dfn=df_groups, dfd=df_errors)
print('critical value =', np.round(cv, 3))

# visualize with plot
x = np.linspace(0, 19, 100)
y = stats.f.pdf(x, df_errors, df_groups)
plt.plot(x, y, label=f'df1 = {df_errors}, df2 = {df_groups}')
plt.axvline(x=f_stat, color='b', label='f statistic')
plt.axvline(x=cv, color='r', label='critical value')

plt.legend();
```

## Feature Selection with ANOVA

```python
# get data
import pandas as pd
from sklearn.model_selection import train_test_split

grades = pd.read_csv('https://raw.githubusercontent.com/gitmystuff/INFO4050/main/Datasets/class-grades.csv', index_col=False)
grades.drop(['Quiz', 'Student', 'Attendance'], axis=1, inplace=True)
print(grades.shape)
print(grades.head())

X_train, X_test, y_train, y_test = train_test_split(grades.drop('FinalGrade', axis=1), grades['FinalGrade'], test_size=.2, random_state=42)

X_train['Assignment'].fillna(X_train['Assignment'].round(decimals=2).median(), inplace=True)
X_train['Tutorial'].fillna(X_train['Tutorial'].round(decimals=2).mode()[0], inplace=True)
X_train['Midterm'].fillna(X_train['Midterm'].round(decimals=2).mean(), inplace=True)
X_train['Final'].fillna(X_train['Final'].round(decimals=2).mean(), inplace=True)
X_train['Activity 2'].fillna(X_train['Activity 2'].round(decimals=2).mean(), inplace=True)

X_test['Assignment'].fillna(X_train['Assignment'].round(decimals=2).median(), inplace=True)
X_test['Tutorial'].fillna(X_train['Tutorial'].round(decimals=2).mode()[0], inplace=True)
X_test['Midterm'].fillna(X_train['Midterm'].round(decimals=2).mean(), inplace=True)
X_test['Final'].fillna(X_train['Final'].round(decimals=2).mean(), inplace=True)
X_test['Activity 2'].fillna(X_test['Activity 2'].round(decimals=2).mean(), inplace=True)
print(X_train.shape, X_test.shape)
print(X_train.info())
```

```python
# identify best features using f_regression and pvalues
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_selection import f_regression
from sklearn.feature_selection import SelectPercentile

model = f_regression(X_train, y_train)
model = pd.Series(model[1])
model.index = X_train.columns
plt.title('pValues sorted from high to low')
model.sort_values(ascending=False).plot.bar();
```

```python
# keep best percentage of features (lowest p values)
keepers = SelectPercentile(f_regression, percentile=60).fit(X_train, y_train)
X_train.columns[keepers.get_support()]
```
