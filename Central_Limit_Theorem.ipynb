# The Central Limit Theorem

The Central Limit Theorem is a fundamental concept in probability theory. It states that if you take a sufficiently large number of samples from a population and calculate the mean of each sample, the distribution of those sample means will approximate a normal distribution (a "bell curve"). This holds true regardless of whether the original population's data is normally distributed.

This theorem is powerful because it explains why the normal distribution appears so frequently in nature and statistics. It allows for the use of statistical methods that assume a normal distribution to be applied to a wide variety of problems, even those involving data with other types of distributions.

**History:** The standard version of the theorem was first proven by the French mathematician Pierre-Simon Laplace in 1810.

**Python Implementation**

The following script first creates a clearly non-normal, right-skewed population. Then, it repeatedly samples from this population, calculates the mean of each sample, and plots the distribution of these means to show that they form a normal distribution.

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set a style for the plots
sns.set_style('whitegrid')

# 1. Create a non-normally distributed population.
# We use a Gamma distribution to create a right-skewed dataset.
# This is important to show that the CLT works even when the original data is not normal.
population_data = np.random.gamma(1, 1000, 5000)
population_mean = np.mean(population_data)

print(f"Original Population Mean: {population_mean:.2f}")

# 2. Create the sampling distribution of the mean.
# We will take 10,000 samples, each of size 100, and calculate the mean of each one.
sample_means = []
num_samples = 10000
sample_size = 100

for i in range(num_samples):
    # Take a random sample from the population.
    sample = np.random.choice(population_data, sample_size)
    # Calculate the mean of the sample and append it to our list.
    sample_means.append(np.mean(sample))

# Calculate the mean of the sample means to show it's close to the population mean.
mean_of_sample_means = np.mean(sample_means)
print(f"Mean of the Sample Means: {mean_of_sample_means:.2f}")

# 3. Visualize the original population and the sampling distribution.
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Plot the original, skewed population distribution.
ax1.set_title("Original Population Distribution (Skewed)")
ax1.set_xlabel("Value")
ax1.set_ylabel("Density")
sns.histplot(population_data, kde=True, ax=ax1, color='blue')
ax1.axvline(population_mean, color='r', linestyle='--', label=f'Population Mean ({population_mean:.2f})')
ax1.legend()

# Plot the distribution of the sample means.
ax2.set_title("Sampling Distribution of the Mean (Normal)")
ax2.set_xlabel("Sample Mean")
ax2.set_ylabel("Density")
sns.histplot(sample_means, kde=True, ax=ax2, color='green')
ax2.axvline(mean_of_sample_means, color='r', linestyle='--', label=f'Mean of Sample Means ({mean_of_sample_means:.2f})')
ax2.legend()

plt.show()
```
