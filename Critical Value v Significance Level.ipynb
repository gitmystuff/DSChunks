# Critical Value v Significance Level

According to Bard

A critical value is a value of a test statistic that separates the rejection region from the acceptance region in a hypothesis test. The level of significance is the probability of rejecting the null hypothesis when it is true.

In other words, the critical value is the value that a test statistic must exceed in order to reject the null hypothesis. The level of significance is the probability of rejecting the null hypothesis when the test statistic exceeds the critical value.

The critical value is determined by the type of test being used and the level of significance. The level of significance is chosen by the researcher.

For example, if a researcher is using a t-test at a significance level of 0.05, the critical value would be the value of the t-statistic that separates the top 5% of the distribution from the bottom 95% of the distribution. If the test statistic exceeds the critical value, the researcher would reject the null hypothesis.

The level of significance is important because it determines how likely it is that the researcher will make a Type I error. A Type I error is the error of rejecting the null hypothesis when it is true.

The level of significance is also important because it affects the power of the test. The power of a test is the probability of rejecting the null hypothesis when it is false.

A higher level of significance will increase the power of the test, but it will also increase the risk of making a Type I error. A lower level of significance will decrease the power of the test, but it will also decrease the risk of making a Type I error.

The researcher must choose a level of significance that is appropriate for the research question being asked.
