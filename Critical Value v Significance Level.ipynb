# Critical Value v Significance Level

## Critical Value and Level
### 1. Key Definitions

| Term | Definition | Role |
| :--- | :--- | :--- |
| **Level of Significance** ($\alpha$) | The probability of **rejecting the null hypothesis ($H_0$) when it is actually true** (committing a **Type I error**). This probability is chosen by the researcher. | Determines the size of the **rejection region** in the sampling distribution. |
| **Critical Value** | A value of the test statistic that **separates the rejection region from the acceptance region** in a hypothesis test. | The **threshold** the calculated test statistic must meet or exceed to reject $H_0$. |

### 2. The Necessary Correction

The second paragraph states: "The level of significance is the probability of rejecting the null hypothesis when the test statistic exceeds the critical value."

* **Correction:** This statement is redundant and slightly misleading. By definition, the **rejection region** *is* the area where the test statistic exceeds the critical value. The **level of significance ($\alpha$)** is the *area (probability)* of that rejection region **under the null hypothesis distribution**.

### 3. The Relationship Illustrated

The Critical Value and the Level of Significance ($\alpha$) are interdependent:

* **Determining the Critical Value:** The critical value is determined by the **type of test** (e.g., $t$-test, $z$-test) and the **level of significance** ($\alpha$) chosen by the researcher.
* **Example (One-Tailed $z$-test, $\alpha = 0.05$):** The critical $z$-value is the point that cuts off the top $5\%$ (or bottom $5\%$) of the $z$-distribution. If the calculated **test statistic** falls into that $5\%$ tail (the rejection region, i.e., it **exceeds the critical value**), the researcher rejects the null hypothesis.

## Importance of the Level of Significance ($\alpha$)

The choice of $\alpha$ is a crucial decision for the researcher because it defines the risk profile of the test:

### 1. Type I Error Risk

The level of significance directly determines the risk of making a **Type I error** (rejecting $H_0$ when it is true). If $\alpha = 0.05$, the researcher accepts a $5\%$ chance of mistakenly concluding that a treatment works when it actually doesn't.

### 2. Power Tradeoff

The level of significance is also related to the **power of the test** (the probability of correctly rejecting $H_0$ when it is false).

* **Higher $\alpha$ (e.g., $0.10$):** **Increases the power** (easier to find an effect) but also **increases the risk of making a Type I error**.
* **Lower $\alpha$ (e.g., $0.01$):** **Decreases the power** (harder to find an effect) but also **decreases the risk of making a Type I error**.

The researcher must select an $\alpha$ value that appropriately balances the risk of a false positive ($\alpha$) against the risk of a false negative (missing a real effect).
